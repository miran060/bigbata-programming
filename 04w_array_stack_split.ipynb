### 난수 기초

```python
import numpy as np

rg = np.random.default_rng(12345)  # 생성자에서 값을주면
rg.random()  # [0, 1)
# random을 붙여도 값은 같게 나오는걸 볼수 있음 / 제너레이터를 만들고 해서 랜덤을 해도 값은 똑같이 나오는 것이다
```

```python
rg.random()
```

```python
rg.random(5) # 하나 쓰면 1차원
```

```python
rg.random( (3, 4)) #2개 쓰면 2차원 배열
```

```python
rg.random( (2, 3, 4)) # 3개 써주면 3차원 배열
```

```python
rg.random(5)
```

```python
import numpy as np

rg = np.random.default_rng(1)
rg.integers(low=0, high=10, size=3) # [0, 10) 정수 난수
  # 항상 값은 똑같음
```

```python
rg.integers(low=0, high=20, size=(3, 4))

# 값은 랜덤이다
```

```python
rg.integers(0, 20, size=(3, 4))

#이것도 랜덤
```

```python
rg.integers(1, 7, (3, 4))

#이것도 랜덤
```

```python
import numpy as np

rg = np.random.default_rng(1)
rg.standard_normal(3) # 표준 정규분포 실수 난수
```

```python
rg.standard_normal((3, 4)) # 표준 정규분포 실수 난수

# 오류가 났던 이유 : ( ( ) ) 이렇게 안하고 () 이거 하나만 적어서
# 이런 오타 많이 나오니 주의 필요......



# ++ 3 넘어가는 건 거의 없고 대부분 2에서 끝남. 95% 확률로
# -9 까지 나온거 확인해봄
```

### 배열 결합 연습

```python
a = np.arange(1, 4)
b = np.arange(4, 7)
c = np.arange(7, 10)
a, b, c
```

```python
p = np.arange(1, 7).reshape(2, 3)
q = np.arange(11, 17).reshape(2, 3)
r = np.arange(21, 27).reshape(2, 3)

p, q, r
```

```python
np.vstack( (p, q, r))
```
