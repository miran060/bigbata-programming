# 다양한 집합 연산

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6, 7}
print(A.union(B))
print(A.intersection(B))
print(A.difference(B))
print(A.symmetric_difference(B))
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6, 7}
print(A | B)
print(A & B)
print(A - B)
print(B - A)
print(A ^ B)
```
