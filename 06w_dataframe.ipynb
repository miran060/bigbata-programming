### 함수 df.query(조건)

```python
df = pd.DataFrame({'A' : range(1, 6),
                   'B' : range(6, -3, -2),
                   'C' : range(10, 5, -1),
                   'D clmns' : range(3, -2, -1),
                   'E':['python', 'java', 'python', 'go', 'python']})
df
```

```python
df.query('A>B')
```

```python
df[df.A > df.B]
```

열 이름을 바로 기술 가능

```python
df.query('C == 10')
```

```python
df.query('A>0')
```

```python
df.query('E == "python"')
```

```python
df.query("E != 'python'")
```

```python
df.query('A == `D clmns` ')
```

```python
df.query('A == " D clmns"')
```

```python
df[df.A == df["D clmns"]]
```

```python
df.query('A>3 | 3 > 8')
```

```python
df.query('A > 2 & C < B')
```

```python
df.query("E in ['python', 'go']")
```

```python
num = 3
df.query('B > @num')
```

```python
num = 3
df.query(f'B > {num}')
```

```python
S = "E.str.contains('py')"
df.query(S)
```

```python
s = 'E.str.contains("B")'
df.query(s)
```

```python
S = "E.str.contains('on')"
df.query(S)
```
